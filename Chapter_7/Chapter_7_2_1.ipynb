### Calculating central tendency measures

Mean

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
mean = np.mean(data)
print(mean)
```

Median

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
median = np.median(data)
print(median)
```

Mode

```python
from scipy import stats
data = np.array([1, 2, 3, 4, 4, 5, 6])
mode = stats.mode(data)[0]
print(mode)
```

Range

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
range = np.ptp(data)
print(range)
```

Interquartile Range (IQR)

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print(IQR)
```

Standard Deviation

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
sd = np.std(data)
print(sd)
```

Variance

```python
import numpy as np
data = np.array([1, 2, 3, 4, 4, 5, 6])
variance = np.var(data)
print(variance)
```

### Skewness and kurtosis in biomedical data

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
np.random.seed(0)  # Creating normally distributed data
sample_data = np.random.normal(0, 1, 1000)  # 1000 data points 
data = pd.DataFrame(sample_data, columns=['Sample Values'])
# Calculate skewness and kurtosis
data_skewness = skew(data['Sample Values'])
data_kurtosis = kurtosis(data['Sample Values'], fisher=False) 
print(f'Skewness: {data_skewness}; Kurtosis: {data_kurtosis}')
```

Distribution Plot

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate the histogram of the sample data
plt.hist(data['Sample Values'], bins=30, density=True, alpha=0.6, color='g')

# Plot the probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data['Sample Values']), np.std(data['Sample Values']))
plt.plot(x, p, 'k', linewidth=2)

title_text = f'Sample Data Distribution\nSkewness: {data_skewness:.2f}, Kurtosis: {data_kurtosis:.2f}'
plt.title(title_text)
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Show the plot with the skewness and kurtosis
plt.show()
```
