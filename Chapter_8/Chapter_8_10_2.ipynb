# 8.10.2. Predicting drug sensitivity using RandomForest

Step 1: Import required libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Data preprocessing

```python
# Load dataset (modify the path as per your dataset)
data = pd.read_csv('example_data/data.csv', index_col=0)

# Split the dataset into features (X) and target variable (y)
X = data.drop('target', axis=1)

y = data['target']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Step 3: Train the Random Forest model

```python
# Initialize the RandomForest Classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
random_forest.fit(X_train, y_train)
```

Step 4: Make predictions and evaluate the model

```python
# Making predictions on the test set
y_pred = random_forest.predict(X_test)

# Evaluating the model
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

Step 5: Visualization of model performance

```python
# Visualizing the confusion matrix
sns.heatmap(conf_matrix, annot=True, fmt='g')
plt.title('Confusion Matrix for RandomForest Model')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```
