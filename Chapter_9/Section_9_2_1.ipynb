## 9.2. Fundamentals of image processing

### 9.2.1. Image representation

```python
import matplotlib.pyplot as plt
from skimage import data, color, filters
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from skimage import io

# Load a sample color image (retina blood vessels image as an example of biomedical data)
color_image = io.imread('example_data/ihc.png')

# Convert the color image to grayscale
gray_image = rgb2gray(color_image)

# Apply Otsu's thresholding to create a binary image from the grayscale image
thresh = threshold_otsu(gray_image)
binary_image = gray_image > thresh  # Binary image

# Display the original color image, grayscale image, and binary image
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Display original color image
ax[0].imshow(color_image)
ax[0].set_title('Color Image')
ax[0].axis('off')

# Display grayscale image
ax[1].imshow(gray_image, cmap='gray')
ax[1].set_title('Grayscale')
ax[1].axis('off')

# Display binary image
ax[2].imshow(binary_image, cmap='gray')
ax[2].set_title('Binary')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
